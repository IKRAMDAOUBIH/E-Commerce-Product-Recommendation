<h1 style="color: #7B68EE; font-weight: bold; font-family: 'Times New Roman';">Méthode de Non Supervisé "unsupervised learning"  :</h1>

<div style="background-color: #F5F0FF; padding: 25px; border-left: 6px solid #7B68EE; border-radius: 0 15px 15px 0; max-width: 1350px;">

<p style="font-family: 'Times New Roman', Times, serif;"><strong style="color: #7B68EE;">Le non supervisé </strong>, ou unsupervised learning est une méthode où le modèle apprend sans étiquettes (labels).</p>

<p style="font-family: 'Times New Roman', Times, serif;">En apprentissage non supervisé, on donne seulement les données (X)Il n’y a pas de variable cible (y) à prédire, Le modèle doit découvrir des structures cachées tout seul.</p>

<p style="font-family: 'Times New Roman', Times, serif; font-weight: bold; color: #7B68EE;">Objectif :</p>

<ul style="font-family: 'Times New Roman', Times, serif; list-style-type: square; color: #5a4a9a;">
Trouver :
    <li style="margin-bottom: 5px;">des groupes (clusters)</li>
    <li style="margin-bottom: 5px;">des relations</li>
    <li style="margin-bottom: 5px;">des patterns cachés</li>
</ul>

</div>

<h1 style="color: #4A90E2; font-weight: bold; font-family: 'Times New Roman';">🔵 Charger les données et le prétraitement  :</h1>

```python
df = pd.read_csv(r"C:\Users\hp\Desktop\ecommerce_project\ecommerce_project\data\Womens Clothing E-Commerce Reviews.csv")
df
```

```python
df = df.dropna(subset=['Review Text'])
```

```python
df['Title'] = df['Title'].fillna("No Title")
```

```python
df['Division Name'] = df['Division Name'].fillna("Unknown")
df['Department Name'] = df['Department Name'].fillna("Unknown")
df['Class Name'] = df['Class Name'].fillna("Unknown")
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
print(df.columns)
```

```python
# Traitement des Outliers (Age)
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(lower, upper)
```

```python
df = df[(df["Age"] >= lower) & (df["Age"] <= upper)]
```

```python
# Traitement des outliers Positive Feedback Count
Q1 = df["Positive Feedback Count"].quantile(0.25)
Q3 = df["Positive Feedback Count"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
df = df[(df["Positive Feedback Count"] >= lower) & 
        (df["Positive Feedback Count"] <= upper)]
```

<h1 style="color: #4A90E2; font-weight: bold; font-family: 'Times New Roman';">🔵 K-Means :</h1>

<div style="position: relative; max-width: 1350px; font-family: 'Times New Roman'; margin-bottom: 20px;">
    <div style="background-color: #7B68EE; padding: 10px 20px 10px 30px; color: white; font-weight: bold; font-size: 1.3em; border-radius: 5px 20px 20px 5px; width: fit-content;">
         Choisir K "méthode du coude":
    </div>
    <div style="position: absolute; left: 0; top: 0; width: 10px; height: 100%; background-color: #4a3a8a; border-radius: 5px 0 0 5px;"></div>
</div>

```python
inertia = []

for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure()
plt.plot(range(1,10), inertia, marker='o')
plt.title("Méthode du coude")
plt.xlabel("Nombre de clusters")
plt.ylabel("Inertia")
plt.show()
```

<div style="font-family: 'Times New Roman';">
    <div style="background: white; padding: 10px 15px; border-radius: 8px; border-left: 4px solid #7B68EE;">
        ➕ Trouver le meilleur nombre de clusters (K) <strong style="color: #7B68EE;">dans ce cas nb = 4 </strong>
    </div>
</div>

<div style="background: linear-gradient(to right, #F5F0FF, #FFFFFF); padding: 20px; border-radius: 10px; max-width: 1350px; font-family: 'Times New Roman'; margin-bottom: 15px; border-left: 6px solid #7B68EE;">

<p style="font-family: 'Times New Roman'; font-size: 1.2em; font-weight: bold; color: #7B68EE; margin-top: 0; margin-bottom: 15px;">
    Méthode du coude - Elbow Method
</p>

<p style="font-family: 'Times New Roman'; color: #2c3e50; margin-bottom: 15px;">
    Méthode <strong style="color: #7B68EE;"> du coude "Elbow Method"</strong> Un graphique en ligne qui trace l'inertie en fonction du nombre de clusters k de 1 à 9. L'objectif est de repérer le "coude", le point où ajouter des clusters ne réduit plus significativement l'inertie.<strong style="color: #7B68EE;"> Résultat :</strong> le coude apparaît à k = 4, ce qui indique que 4 clusters est le nombre optimal pour partitionner ce dataset avec K-Means.
</p>

</div>

<div style="position: relative; max-width: 1350px; font-family: 'Times New Roman'; margin-bottom: 20px;">
    <div style="background-color: #7B68EE; padding: 10px 20px 10px 30px; color: white; font-weight: bold; font-size: 1.3em; border-radius: 5px 20px 20px 5px; width: fit-content;">
         Appliquer K-Means:
    </div>
    <div style="position: absolute; left: 0; top: 0; width: 10px; height: 100%; background-color: #4a3a8a; border-radius: 5px 0 0 5px;"></div>
</div>

```python
kmeans = KMeans(n_clusters=4, random_state=42)
labels_kmeans = kmeans.fit_predict(X_scaled)
```

<div style="position: relative; max-width: 1350px; font-family: 'Times New Roman'; margin-bottom: 20px;">

```python
centroids = kmeans.cluster_centers_
print(centroids)
```

<h1 style="color: #4A90E2; font-weight: bold; font-family: 'Times New Roman';">🔵 DBSCAN:</h1>

<div style="background: linear-gradient(to right, #F5F0FF, #FFFFFF); padding: 20px; border-radius: 10px; max-width: 1350px; font-family: 'Times New Roman'; margin-bottom: 15px; border-left: 6px solid #7B68EE;">

<p style="font-family: 'Times New Roman'; font-size: 1.2em; font-weight: bold; color: #7B68EE; margin-top: 0; margin-bottom: 15px;">
    DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
</p>

<p style="font-family: 'Times New Roman'; color: #2c3e50; margin-bottom: 15px;">
    DBSCAN est un algorithme de clustering basé sur la <strong style="color: #7B68EE;">densité des données</strong>.
</p>

<ul style="font-family: 'Times New Roman'; color: #5a4a9a; margin-bottom: 15px;">
    <li style="margin-bottom: 8px;">Il regroupe les <strong style="color: #7B68EE;">points proches</strong> et identifie les points isolés comme <strong style="color: #7B68EE;">bruit (-1)</strong>.</li>
    <li style="margin-bottom: 8px;">Son rôle est de découvrir des <strong style="color: #7B68EE;">groupes naturels</strong> et détecter les <strong style="color: #7B68EE;">outliers</strong> sans définir le nombre de clusters.</li>
</ul>

</div>

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(X_scaled)
```

<div style="background: linear-gradient(to right, #F5F0FF, #FFFFFF); padding: 20px; border-radius: 10px; max-width: 1350px; font-family: 'Times New Roman'; margin-bottom: 15px; border-left: 6px solid #7B68EE;">

<p style="font-family: 'Times New Roman'; font-size: 1.2em; font-weight: bold; color: #7B68EE; margin-top: 0; margin-bottom: 15px;">
    Paramètres DBSCAN
</p>

<div style="display: flex; gap: 12px; flex-wrap: wrap; margin-bottom: 15px;">
    <div style="background: white; padding: 8px 16px; border-radius: 30px; border: 1px solid #7B68EE;">
        <span style="font-weight: bold; color: #7B68EE;">eps</span>
        <span style="color: #5a4a9a;"> → distance max</span>
    </div>
    <div style="background: white; padding: 8px 16px; border-radius: 30px; border: 1px solid #7B68EE;">
        <span style="font-weight: bold; color: #7B68EE;">min_samples</span>
        <span style="color: #5a4a9a;"> → min points pour cluster</span>
    </div>
</div>

<div style="background: #FF6B6B10; padding: 10px 15px; border-radius: 8px;">
    <span style="color: #FF6B6B;">⚠️</span> <span style="color: #2c3e50;">Peut détecter les anomalies <strong style="color: #FF6B6B;">(-1)</strong></span>
</div>

</div>

<div style="position: relative; max-width: 1350px; font-family: 'Times New Roman'; margin-bottom: 20px;">
    <div style="background-color: #7B68EE; padding: 10px 20px 10px 30px; color: white; font-weight: bold; font-size: 1.3em; border-radius: 5px 20px 20px 5px; width: fit-content;">
         Visualisation DBSCAN :
    </div>
    <div style="position: absolute; left: 0; top: 0; width: 10px; height: 100%; background-color: #4a3a8a; border-radius: 5px 0 0 5px;"></div>
</div>

```python
plt.figure()
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels_dbscan)
plt.title("DBSCAN Clustering")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```
